```python
#Lists
names = ["john", "Bob", "Mosh", "Sam", "Mary"]
print(names)
print(names[0])
print(names[-1])
print(names[-2])

#updating list objects
names[0] = "Jon"
print(names)

#Printing indexes from 0 to 2
print(names[0:3]) #3 index is excluded
print(names)

#checking elements
print("Jon" in names)

#checking number of elements in the list
print(len(names))

#to add an element at the end of the list
names.append("Faiyaz")
print(names)

#to add an element at any index position of the list
names.insert(0,"Abrar") #insert(index number, value)
print(names)

#to remove object from the list
names.remove("Abrar")
print(names)

#to make the list clear/empty
names.clear()
print(names)
```

```python
#loops
numbers = [1, 2, 3, 4, 5]

#for loop
for item in numbers:
    print(item)

#while loop
i = 1
while i <= len(numbers):
    print(i)
    i += 1
```

```python
#range object
numbers = range(5) #generates numbers from 0 upto the number we specify
print(numbers)
for number in numbers:
    print(number)

numbers = range(5,10) #generates numbers from 5 upto the number we specify
print(numbers)

for number in numbers:
    print(number)

numbers = range(5,10,2) #generates numbers from 0 upto the number we specify with skipping 2 steps at a time
print(numbers)

for number in numbers:
    print(number)

#Appropiate way to use range function
for number in range(5,10):
    print(number)
```

```python
#Tuple
#tuples are immutable/unchangable
#which means we can not assign or remove any element from tuple
numbers = (1,2,3,3,2)

print(numbers.count(3))

print(numbers.index(2))
```

```python
#Dictionaries
#Dictionaries can not have two items with the same key.
#it will not show any error.instead it will take the last value of the same key.
#can take any types of elements
thisdict = {
    "brand" : "Ford",
    "model" : "Mustang",
    "electric" : False,
    "year" : 1964
}
print(thisdict)

print(thisdict["brand"])

thisdict["year"] = 1969 #updating values
print(thisdict)

thisdict["color"] = ["red", "blue" , "white"] #inserting
print(thisdict)

thisdict.pop("model") #deleting item
print(thisdict)

thisdict.popitem() #deleting the last item
print(thisdict)

print(thisdict.keys())

print(thisdict.values())

print(thisdict.items()) #returns items as pair of tuples
print(len(thisdict))

#copying dictionary
mydict = thisdict.copy()
print(mydict)

#dictionaries inside dictionary
myfamily = {
    "child1" : {
        "name" : "Eva",
        "year" : 2004
    },
    "child2" : {
        "name" : "Ivy",
        "year" : 2000
    },
    "child3" : {
        "name" : "Labiba",
        "year" : 2000
    },
    "child4" : {
        "name" : "Nafisa",
        "year" : 2001
    }
}
print(myfamily)

print(thisdict.clear()) #making the dictionary empty
```
